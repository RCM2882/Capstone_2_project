```python
df_cleaned['month'] = df_cleaned['DateAdded'].dt.month
df_cleaned['year'] = df_cleaned['DateAdded'].dt.year
```

```python
monthly_reviews = df_cleaned.groupby(['year', 'month']).agg({'NumReviews': 'sum'}).reset_index()
monthly_reviews['date'] = pd.to_datetime(monthly_reviews[['year', 'month']].assign(day=1))
```

```python
# Number of reviews over time

plt.figure(figsize=(14, 8))
plt.plot(monthly_reviews['date'], monthly_reviews['NumReviews'], marker='o')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.show()
```

```python
# Seasonal Patterns
seasonal_reviews = df_cleaned.groupby('month').agg({'NumReviews': 'sum'}).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='NumReviews', data=seasonal_reviews, palette='viridis')
plt.title('Seasonal Patterns in Number of Reviews')
plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
monthly_prices = df_cleaned.groupby(['year', 'month']).agg({'Price': 'mean'}).reset_index()
monthly_prices['date'] = pd.to_datetime(monthly_prices[['year', 'month']].assign(day=1))
```

```python
# Average price trends over time

plt.figure(figsize=(14, 8))
plt.plot(monthly_prices['date'], monthly_prices['Price'], marker='o', linestyle='-')
plt.title('Average Price Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

```python
seasonal_prices = df_cleaned.groupby('month').agg({'Price': 'mean'}).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='month' , y='Price', data=seasonal_prices, palette='viridis')
plt.title('Seasonal Trends in Average Prices')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
# Pearson correlation coefficient between rating and number of reviews
correlation = df_cleaned['Rating'].corr(df_cleaned['NumReviews'])
print(f'Pearson correlation coefficient between rating and number of reviews: {correlation}')
```

```python
category_correlations = df_cleaned.groupby('Category').apply(lambda x: x['Rating'].corr(x['NumReviews']))
category_correlations = category_correlations.reset_index()
category_correlations.columns = ['Category', 'Correlation']
print(category_correlations)
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='Correlation', y='Category', data=category_correlations, palette='coolwarm')
plt.title('Pearson Correlation Coefficent between Rating and Number of Reviews by Category')
plt.xlabel('Pearson Correlation Coefficent')
plt.ylabel('Category')
plt.show()
```
